```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from lifelines import WeibullFitter, ExponentialFitter, LogNormalFitter,LogLogisticFitter, WeibullAFTFitter, \
LogNormalAFTFitter, LogLogisticAFTFitter
```

## Reading the data

```python
df = pd.read_csv('telco.csv')
df
```

## Data manipulation

### Index

```python
df.set_index('ID', inplace=True)
```

### Churn to binary

```python
df['churn'] = pd.Series(np.where(df.churn.values == "Yes", 1, 0), df.index)
```

### Categorical to dummy

```python
encode_cols = ['region', 'marital', 'ed', 'retire', 'gender', 'voice', 'internet', 'forward', 'custcat'] 
survival_df = pd.get_dummies(df, columns=encode_cols, prefix=encode_cols, drop_first=True)
```

### Zeros to small floats

```python
survival_df["tenure"] = np.where(survival_df["tenure"] == 0, 0.000001, survival_df["tenure"])
```

## Prepared data

```python
survival_df
```

## Parametric models

### Building different models

```python
wf = WeibullFitter()
ef = ExponentialFitter()
lnf = LogNormalFitter()
llf = LogLogisticFitter()
```

### Fitting

```python
for model in [wf, ef, lnf, llf]:
    model.fit(durations = survival_df["tenure"], event_observed = survival_df["churn"])
    print("The BIC of", model.__class__.__name__, "is",  model.BIC_) 
    model.print_summary()
    print('='*100)
```

### Plotting

```python
for model in [wf, ef, lnf, llf]:
    model.fit(durations = survival_df["tenure"], event_observed = survival_df["churn"])
    model.plot_survival_function()
```

## ATF

```python
wf_aft = WeibullAFTFitter()
lnf_aft = LogNormalAFTFitter()
llf_aft = LogLogisticAFTFitter()
```

### Fitting

```python
for model in [wf_aft, lnf_aft, llf_aft]:
    model.fit(survival_df, duration_col = "tenure", event_col = "churn")  
    print("The AIC value for", model.__class__.__name__, "is",  model.AIC_)
    print("The BIC of", model.__class__.__name__, "is",  model.BIC_) 
    model.print_summary()
    print('='*100)
```

### Plotting

```python
for model in [wf_aft, lnf_aft, llf_aft]:
    plt.plot(model.predict_survival_function(survival_df.loc[1]), label = model.__class__.__name__)
    plt.legend()
```

## Choosing a model

While the models perform similarly, LogNormalAFTFitter has the lowest AIC and BIC scores and thus is a good candidate to pick.

## Choosing features

```python
survival_df = survival_df[["tenure", "churn", "address", "age", "custcat_E-service", "custcat_Plus service", \
                           "custcat_Total service", "internet_Yes", "marital_Unmarried", "voice_Yes"]]
```

## The final model

```python
logn_aft = LogNormalAFTFitter()
logn_aft.fit(survival_df, duration_col='tenure', event_col='churn')
logn_aft.print_summary()
```

## CLV

```python
pred_clv = logn_aft.predict_survival_function(survival_df)
pred_clv
```

```python
pred = pred_clv.loc[1:12, :]
pred
```

```python
MM = 1300
r = 12

for col in range(1, len(pred.columns) + 1):
    for row in range(1, 13):
        pred.loc[row, col] = pred.loc[row, col] / (1 + r / 12) ** (row - 1)

df['CLV'] = MM * pred.sum(axis=0)

df
```

```python
clv = sns.displot(df, x = 'CLV', kind = 'kde', hue = 'custcat')
clv.set(title = 'Customer Category CLV density')
```

Plus service is the most likely to survive.

```python
# within a year
average_clv_left = df["CLV"].loc[(df["tenure"] <= 12) & (df["churn"] == 1)].mean()
print(average_clv_left)

# in general
clv_yearly = df["CLV"].loc[(df["tenure"] <= 12) & (df["churn"] == 1)].sum()
print(clv_yearly)
```

## Report

### Interpretation of Coefficients

The Log-Normal Fitter model highlighted that tenure, age, education, internet usage, and customer category significantly influence churn risk.


For instance, customers with Total Service, E-service and Basic Service subscription types had higher hazard ratios, indicating an increased risk of churn.

### Valuable Segments

Valuable segments were defined based on the model coefficients. Customers with Total Service, E-service, and Plus Service subscriptions, younger age, and higher tenure were identified as potentially more valuable, even though at a higher churn risk. These segments may require targeted retention strategies.

### Annual Retention Budget

To estimate the annual retention budget, factors like CLV and survival probabilities should be considered. The survival analysis models provide insights into customer tenure, aiding in the identification of at-risk subscribers within a year. Calculating the budget would require a detailed CLV calculation, factoring in acquisition costs and the predicted survival probabilities.

### Retention Strategies

Additional strategies for retention could involve:
1. **Customer Engagement Programs:** Implementing programs to enhance customer engagement, satisfaction, and loyalty.
2. **Personalized Offerings:** Tailoring offers and services based on individual preferences and behaviors.
3. **Feedback Mechanisms:** Establishing mechanisms to collect and act upon customer feedback for continuous improvement.
